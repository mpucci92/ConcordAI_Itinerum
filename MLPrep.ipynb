```python
import pandas as pd
import numpy as np

import sys, os

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('features.csv')
```

```python
sns.heatmap(df.notnull())
```

```python
df.fillna(value=0, inplace=True)
```

```python
sns.heatmap(df.notnull())
```

```python
df.describe()
```

```python
def fill_it(x):
    if x == np.nan or x == np.inf or x == -np.inf:
        return 0
    else:
        return x
```

```python
df.MaxAccel = df.MaxAccel.apply(fill_it)
```

```python
df.MeanAccel = df.MeanAccel.apply(fill_it)
```

```python
df.describe()
```

### Scaling

```python
df.DuplcateLL = np.log(df.DuplcateLL+1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
features_to_scale = ['SpeedPct', 'MaxSpeed', 'STDSpeed', 'MeanSpeed', 'AccelPct', 'MaxAccel', 'STDAccel', 'MeanAccel', 'TotalDistance', 'DuplcateLL']
```

```python
ss = StandardScaler()
df.loc[:, features_to_scale] = ss.fit_transform(df[features_to_scale])
```

### Logging

```python
features_to_ohe = ['Bixies', 'RushType', 'City']
```

```python
df = pd.get_dummies(df, columns = features_to_ohe)
```

```python
df.describe()
```

### Features Done

```python
modes = np.load('modes.npy')
modes = pd.DataFrame(modes, columns=['ID', 'Mode'])
```

```python
modes.ID = modes.ID.astype(int)
```

```python
df = df.merge(modes, on='ID', how='outer').dropna()
```
